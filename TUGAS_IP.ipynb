```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgA = cv2.imread('/content/A.jpg', cv2.IMREAD_GRAYSCALE)
imgB = cv2.imread('/content/B.jpg', cv2.IMREAD_GRAYSCALE)

print("Ukuran A:", imgA.shape)
print("Ukuran B:", imgB.shape)
```

```python
if imgA.shape != imgB.shape:
    imgB = cv2.resize(imgB, (imgA.shape[1], imgA.shape[0]))
    print("Ukuran B disamakan:", imgB.shape)
```

```python
M, N = imgA.shape
imgC = np.zeros((M, N), dtype=np.uint8)

for i in range(M):
    for j in range(N):
        temp = int(imgA[i, j]) + int(imgB[i, j])
        if temp > 255:
            imgC[i, j] = 255
        else:
            imgC[i, j] = temp
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Citra A")
plt.imshow(imgA, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Citra B")
plt.imshow(imgB, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Citra C = A + B")
plt.imshow(imgC, cmap='gray')
plt.axis('off')

plt.show()
```

```python
cv2.imwrite('/content/hasil_penjumlahan.png', imgC)
print("Tersimpan: /content/hasil_penjumlahan.png")
```
